# DATOS DE OBRAS EXPUESTAS DE AMBOS MUSEOS

```python
# Datos del Museo del Prado
prado_expuestos = 1136
prado_expuestas = 14

# Datos del Museo Reina Sofía
reina_sofia_expuestos = 333
reina_sofia_expuestas = 36
```

```python
# Etiquetas 
museos = ['Museo del Prado', 'Museo Reina Sofía']

# Alturas de las barras
hombres = [prado_expuestos, reina_sofia_expuestos]
mujeres = [prado_expuestas, reina_sofia_expuestas]
# Posición de las barras
posicion = range(len(museos))
```

# OCUPACIÓN

```python
empleos.ocupacion #filtro por la ocupación
```

```python
def ocupaciones(x):
    if type(x) == str:
        x= x.strip()
        x = x.split('|')
        return x
    else:
        return []
listas_empleos = empleos.ocupacion.map(ocupaciones)
listas_empleos
```

```python
#creo una nueva columna llamada listas_empleos. Devuelve una lista de listas con las ocupaciones divididas
listas_empleos = empleos.ocupacion.map(ocupaciones)
listas_empleos
```

```python
# creO una lista bidimensional y después fusiono todas las sublistas en una sola lista 
import itertools
list2d = listas_empleos.to_list()
merged = pd.Series(list(itertools.chain(*list2d)), name = 'Profesiones')
merged
```

```python
# cuento los valores únicos en la Serie 
profesiones = merged.value_counts()
profesiones
```

```python
profesiones[:30]
```

```python
# Gráfico de barras de todas las ocupaciones
plt.figure(figsize=(10, 10))
profesiones[:10].plot(kind = 'barh', color='#B25068')
plt.title('Número de mujeres por ocupación')
plt.show()
```
