# PORCENTAJE DE VISITANTES DE LOS MUSEOS ESPAÑOLES

# Extraigo  los valores de los visitantes en la fila 3 y 4 de la columna 'Column2' y asigno a las variables 'hombres' y 'mujeres'. Creo una lista llamada 'datos' con los valores de 'hombres' y 'mujeres', y otra lista llamada 'etiquetas' con las etiquetas 'Hombres' y 'Mujeres'.

```python
hombres = visitantes.loc[3, 'Column2']
mujeres = visitantes.loc[4, 'Column2']
datos = [hombres, mujeres]
etiquetas = ['Hombres', 'Mujeres']
colores = ['#B25068', '#E7AB79']
```

```python

plt.pie(datos, colors=colores, autopct='%1.1f%%')
plt.axis('equal')  
plt.legend(etiquetas)
plt.show()
```

# VISITANTES POR RANGOS DE EDAD

```python
personas_1519 = visitantes.loc[5, 'Column2']
personas_2024 = visitantes.loc[6, 'Column2']
personas_2534 = visitantes.loc[7, 'Column2']
personas_3544 = visitantes.loc[8, 'Column2']
personas_4554 = visitantes.loc[9, 'Column2']
personas_5564 = visitantes.loc[10, 'Column2']
personas_6574 = visitantes.loc[11, 'Column2']
personas_75 = visitantes.loc[12, 'Column2']
```

```python
datos = [personas_1519,personas_2024,personas_2534,personas_3544,personas_4554,personas_5564,personas_6574,personas_75]
etiquetas = ['15-19', ' 20-24', '25-34', '35-44', '45-54', '55-64', '65-74', 'Más de 75']
```

# Extraigo los alores de los visitantes en diferentes grupos de edad en las filas 5 a 12 de la columna 'Column2' y asigno a variables separadas. Luego, creo una lista llamada 'datos' con los valores de cada grupo de edad y otra lista llamada 'etiquetas' con las etiquetas correspondientes a cada grupo de edad.

```python
# Inverto los datos y etiquetas
datos = datos[::-1]
etiquetas = etiquetas[::-1]
# Creo el gráfico 
plt.bar(etiquetas, datos, color='#E7AB79')

# Personalizo el gráfico
plt.xlabel('Rango de edad')
plt.ylabel('Porcentaje de visitantes')
plt.show()
```
