# DATOS DE OBRAS EXPUESTAS DE AMBOS MUSEOS

```python
# Datos del Museo del Prado
prado_expuestos = 1136
prado_expuestas = 14

# Datos del Museo Reina Sofía
reina_sofia_expuestos = 333
reina_sofia_expuestas = 36
```

```python
# Etiquetas de los museos
museos = ['Museo del Prado', 'Museo Reina Sofía']

# Alturas de las barras
hombres = [prado_expuestos, reina_sofia_expuestos]
mujeres = [prado_expuestas, reina_sofia_expuestas]
# Posición de las barras
posicion = range(len(museos))
```

```python
# Crear la figura y los ejes
fig, ax = plt.subplots()
# Crear las barras
bar1 = ax.bar(posicion, hombres, color='#9b59b6', label='Hombres')
bar2 = ax.bar(posicion, mujeres, color='#ebdef0', label='Mujeres', bottom=hombres)

# Configurar los ejes y la leyenda
ax.set_ylabel('Número de obras')
ax.set_title('Obras expuestas por género en los museos')
ax.set_xticks(posicion)
ax.set_xticklabels(museos)
ax.legend()

# Mostrar el gráfico
plt.show()
```

# REPRESENTACIÓN POR PAÍSES

```python
empleos.nacPais.unique()
```

```python
paises = empleos.nacPais.value_counts()

# paises con más de 20 artistas
mas_de_20_artistas = paises > 20
paises_mas20 = paises[mas_de_20_artistas]

# paises con menos de 20 artistas
menos_de_20_artistas = paises <= 20
paises_menos20 = paises[menos_de_20_artistas]
```

```python

# Grafica resultados (paises )
paises_mas20[:10].plot(kind='bar', color='#9b59b6')
plt.xlabel('País')
plt.ylabel('Cantidad de mujeres')
#plt.title('Mujeres en el Museo Reina Sofía por país \n Países con más de 20 artistas')
plt.show()
```

```python

# Grafica resultados (paises )
paises_menos20[:10].plot(kind='bar', color='#9b59b6' )
plt.xlabel('País')
plt.ylabel('Cantidad de mujeres')
#plt.title('Mujeres en el Museo Reina Sofía por país \n Países con menos de 20 artistas')
plt.show()
```

# OCUPACIÓN

```python
empleos.ocupacion
```

```python
def ocupaciones(x):
    if type(x) == str:
        x= x.strip()
        x = x.split('|')
        return x
    else:
        return []
listas_empleos = empleos.ocupacion.map(ocupaciones)
listas_empleos
```

```python
import itertools
list2d = listas_empleos.to_list()
merged = pd.Series(list(itertools.chain(*list2d)), name = 'Profesiones')
merged
```

```python
profesiones = merged.value_counts()
profesiones
```

```python
profesiones[:30]
```

```python
# Gráfico de barras de todas las ocupaciones
plt.figure(figsize=(10, 10))
profesiones[:10].plot(kind = 'barh', color='#9b59b6')
plt.title('Número de mujeres por ocupación')
plt.show()
```
