# Representación por países 

```python
# Genial . Aquí si que tenemos los países
empleos.nacPais.unique()
```

```python
paises = empleos.nacPais.value_counts()

# paises con más de 20 artistas
mas_de_20_artistas = paises > 20
paises_mas20 = paises[mas_de_20_artistas]

# paises con menos de 20 artistas
menos_de_20_artistas = paises <= 20
paises_menos20 = paises[menos_de_20_artistas]
```

```python

# Grafica resultados (paises )
paises_mas20.plot(kind='bar', color='#FCD9BC')
plt.xlabel('País')
plt.ylabel('Cantidad de mujeres')
plt.title('Mujeres en el Museo Reina Sofía por país \n Países con más de 20 artistas')
plt.show()
```

```python

# Grafica resultados (paises )
paises_menos20.plot(kind='barh', figsize = (7, 12) )
plt.xlabel('País')
plt.ylabel('Cantidad de mujeres')
plt.title('Mujeres en el Museo Reina Sofía por país \n Países con menos de 20 artistas')
plt.show()
```

```python
pib = pd.read_excel('PIB_2021.xlsx')
def quitar_espacios_importes(x):
    l = x.split()
    d = ''.join(x for x in l)    
    return int(d)


def quitar_espacios_nombres_pais(x):
    l = x.strip()    
    return l


pib['PIB (float)'] = pib['PIB (nominal millones de USD)'].map(quitar_espacios_importes)
pib.País = pib.País.map(quitar_espacios_nombres_pais)
pib
```

```python
## Cruzamos los datos con los paises del museo
```

```python
paises
```

```python
pib.País.value_counts()
```

```python
pib_numero_mujers = pd.merge(paises, pib, left_index = True, right_on = 'País' )
pib_numero_mujers
```

```python
import plotly.graph_objs as go

from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
trace = go.Scatter(
    x = pib_numero_mujers.nacPais,
    y = pib_numero_mujers['PIB (float)'],
    mode = 'markers',
    text = pib_numero_mujers.País,
    
)

data = [trace]

layout = dict(
            title='Número de artistas mujeres por país ',
            yaxis = dict(
                title = "PIB (en USD)"
            ),
            xaxis = dict(
                title = "Número de mujeres artistas"
            )
         )
fig = go.Figure(data=data, layout=layout)
iplot(fig)

```

No parece que el número de artistas aumente si aumenta el pib del país.

# Ocupación

```python
empleos.ocupacion
```

```python
def ocupaciones(x):
    if type(x) == str:
        x= x.strip()
        x = x.split('|')
        return x
    else:
        return []
listas_empleos = empleos.ocupacion.map(ocupaciones)
listas_empleos
```

```python
import itertools
list2d = listas_empleos.to_list()
merged = pd.Series(list(itertools.chain(*list2d)), name = 'Profesiones')
merged
```

```python
profesiones = merged.value_counts()
profesiones
```

```python
profesiones[:30]
```

```python
# Gráfico de barras de todas las ocupaciones
plt.figure(figsize=(10, 10))
profesiones[:50].plot(kind = 'barh')
plt.title('Número de mujeres por ocupación')
plt.show()
```

#Datos de museos


```python
# Datos del Museo del Prado
prado_hombres = 6420
prado_mujeres = 54

# Datos del Museo Reina Sofía
reina_sofia_hombres = 4665
reina_sofia_mujeres = 658
```

```python
# Etiquetas de los museos
museos = ['Museo del Prado', 'Museo Reina Sofía']
```

```python

# Alturas de las barras
hombres = [prado_hombres, reina_sofia_hombres]
mujeres = [prado_mujeres, reina_sofia_mujeres]
```

```python
# Posición de las barras
posicion = range(len(museos))


```

```python
# Crear la figura y los ejes
fig, ax = plt.subplots()
# Crear las barras
bar1 = ax.bar(posicion, hombres, color='#B68B67', label='Hombres')
bar2 = ax.bar(posicion, mujeres, color='#FCD9BC', label='Mujeres', bottom=hombres)

# Configurar los ejes y la leyenda
ax.set_ylabel('Número de pinturas')
ax.set_title('Distribución de género en los museos')
ax.set_xticks(posicion)
ax.set_xticklabels(museos)
ax.legend()

# Mostrar el gráfico
plt.show()
```
